### Задача 4. Определение целевой переменной в данных и распределения переменных
**Что нужно сделать**

1. Исходя из поставленной задачи определите, в какой колонке находится целевая переменная.



**Ответ:**  Целевая переменная находится в колонке price_category.

2. Выведите для каждого значения целевой переменной его долю вхождения в выборку.


````python
target_column = 'price_category' # Определение целевой переменной
value_counts = df[target_column].value_counts(normalize=True) # Вычисление доли вхождения каждого значения целевой переменной
````

````python
print(f"\nДоля вхождения каждого значения целевой переменной '{target_column}':")
print(value_counts)
````

3. Ещё раз посмотрите на долю или количество вхождений каждого значения целевой переменной в выборку и напишите, какое это распределение. Определить, как распределены значения целевой переменной, в данном случае можно и без графика. 

**Ответ:** Распределение значений целевой переменной price_category является почти равномерным. Все три категории (high, medium, low) представлены в датасете примерно в равных долях, без явного доминирования одной категории над другими.

````python
plt.figure(figsize=(8, 6))
sns.countplot(x=target_column, data=df, order=df[target_column].value_counts().index)
plt.title('Распределение значений целевой переменной price_category')
plt.xlabel('Категории цен')
plt.ylabel('Количество записей')
plt.show()
````
